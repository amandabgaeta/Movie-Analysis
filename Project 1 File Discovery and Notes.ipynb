```python
#fill in studio nulls
bom_mg_df['studio'].isna().sum()
bom_mg_df['studio'] = bom_mg_df['studio'].fillna('Unknown')
bom_mg_df['studio'].isna().sum()
```

```python
imdb_tb_df.info()
```

```python
# add a concatenated column of title and year to match by in both tables
# first to bom_mg_df
bom_mg_df['title_year'] = bom_mg_df['title'] + ' ' + bom_mg_df['year'].astype(str)
bom_mg_df.head()
```

```python
# then create column in imdb_tb_df
imdb_tb_df['title_year'] = imdb_tb_df['primary_title'] + ' ' + imdb_tb_df['start_year'].astype(str)
imdb_tb_df.head()
```

```python
# new merged data set based on concatenated string
mg_tb = bom_mg_df.merge(imdb_tb_df, on='title_year')
mg_tb.info()
```

```python
#convert foreign_gross from object to float (or both grosses to int so can add for total)
mg_tb['foreign_gross'] = mg_tb['foreign_gross'].str.replace(',','').astype(float)
```

```python
#add columns that have foreign_gross and domestic_gross converted to in millions
mg_tb['foreign_gross_in_mil'] = (mg_tb['foreign_gross']/1000000)
mg_tb['domestic_gross_in_mil'] = (mg_tb['domestic_gross']/1000000)
mg_tb.head()
```

```python
#add column for total_gross in millions
mg_tb['total_gross_in_mil'] = mg_tb['domestic_gross_in_mil'] + mg_tb['foreign_gross_in_mil']
mg_tb.head()
```

```python
#add columns that have foreign_gross and domestic_gross as % of total_gross
mg_tb['domestic_gross_p'] = round((mg_tb['domestic_gross_in_mil']/mg_tb['total_gross_in_mil']),2)
mg_tb['foreign_gross_p'] = round((mg_tb['foreign_gross_in_mil'][0:20]/mg_tb['total_gross_in_mil'][0:20]),2)
mg_tb.head()
```

```python
# understand overview of gross amounts and %s for further analysis/digging into specifics

print(mg_tb['total_gross_in_mil'].describe()) #mean twice than median, large outliers/blockbuster movies; large std
print(mg_tb['domestic_gross_p'].describe()) #avg 44% domestic
print(mg_tb['foreign_gross_p'].describe()) # avg 60-62% international
```
